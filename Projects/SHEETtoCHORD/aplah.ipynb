```python
import re
from collections import Counter
import music21

class MusicAnalyzer:
    def __init__(self, musicxml_file):
        self.musicxml_file = musicxml_file
        self.score = None
        self.key_signature = None
        self.detected_key = None
        self.bars_notes = []
        self.bars_notes_objs = []
        self.all_notes = []

        # Weight configurations (tunable parameters)
        self.weights = {
            'base_match': 1.0,
            'duration_factor': 0.5,
            'bass_match': 5.0,
            'root_in_bass': 4.0,
            'fifth_in_bass': 2.0,
            'third_in_bass': 1.5,
            'top_note_root': 2.0,
            'top_note_third': 1.5,
            'top_note_fifth': 1.2,
            'short_note_penalty': 0.4,
            'cadence_5_1': 8.0,
            'cadence_4_1': 5.0,
            'first_bar_tonic': 10.0,
            'last_bar_tonic': 12.0,
        }
        self.note_cache = {}  # Cache for normalized note names

        self.load_musicxml()

    # Loading & extraction
    def load_musicxml(self):
        try:
            print(f"Loading MusicXML file: {self.musicxml_file}")
            self.score = music21.converter.parse(self.musicxml_file)

            # Get key from file
            key_elements = self.score.flat.getElementsByClass(music21.key.Key)
            if key_elements:
                self.key_signature = str(key_elements[0])
            else:
                key_sig_elements = self.score.flat.getElementsByClass(music21.key.KeySignature)
                if key_sig_elements:
                    self.key_signature = str(key_sig_elements[0])
                else:
                    self.key_signature = None

            self.extract_notes()

            print(f"Successfully loaded file. Found {len(self.bars_notes)} bars.")
        except Exception as e:
            print(f"Error loading MusicXML file: {e}")
            raise

    def normalize_note_name(self, note_name):
        if not note_name:
            return None

        # Check cache first
        if note_name in self.note_cache:
            return self.note_cache[note_name]

        # Remove octave numbers
        name = re.sub(r'\d', '', note_name)

        # Normalize unicode accidentals
        name = name.replace("♯", "#").replace("♭", "b").replace("♮", "")

        # Expanded enharmonic conversion (sharps preferred)
        ENH = {
            # Flats to sharps
            "Cb": "B", "B#": "C",
            "Db": "C#", "C##": "D", "Dbb": "C",
            "Eb": "D#", "D##": "E", "Ebb": "D",
            "Fb": "E", "E#": "F",
            "Gb": "F#", "F##": "G", "Gbb": "F",
            "Ab": "G#", "G##": "A", "Abb": "G",
            "Bb": "A#", "A##": "B", "Bbb": "A",
            # Double flats/sharps
            "Cbb": "Bb", "B##": "C#",
            "Dbb": "C", "Cx": "D",  # C## = Cx
            "Ebb": "D", "Dx": "E",
            "Fbb": "Eb", "Ex": "F#",
            "Gbb": "F", "Fx": "G",
            "Abb": "G", "Gx": "A",
            "Bbb": "A", "Ax": "B",
            # Handle E# and B# properly
            "E#": "F", "B#": "C",
        }

        # If already natural
        if name in ["A", "B", "C", "D", "E", "F", "G"]:
            self.note_cache[note_name] = name
            return name

        # Convert single accidental notes
        if name in ENH:
            result = ENH[name]
            self.note_cache[note_name] = result
            return result

        # Edge-case: unexpected patterns → try stripping unknown chars
        base = re.sub(r'[^A-G#b]', '', name)
        result = ENH.get(base, base)
        self.note_cache[note_name] = result
        return result

    def extract_notes(self):
        """
        Extract notes from all parts and store duration + pitch + type.
        """
        self.all_notes = []
        self.bars_notes = []
        self.bars_notes_objs = []

        if not self.score.parts:
            return

        parts_measures = []
        max_measure_number = 0

        # Collect all measures
        for part in self.score.parts:
            pm = list(part.getElementsByClass(music21.stream.Measure))
            parts_measures.append(pm)
            max_measure_number = max(max_measure_number, len(pm))

        # For each measure
        for i in range(max_measure_number):
            merged_notes = []
            merged_objs = []
            measure_has_notes = False

            for part_measure_list in parts_measures:
                if i < len(part_measure_list):
                    measure = part_measure_list[i]

                    # Check if measure is empty or has only rests
                    if not measure.notesAndRests:
                        continue

                    for el in measure.notesAndRests:
                        # Skip rests
                        if isinstance(el, music21.note.Rest):
                            continue

                        # Single NOTE
                        if isinstance(el, music21.note.Note):
                            entry = {
                                "name": self.normalize_note_name(el.pitch.name),
                                "duration": float(el.quarterLength),
                                "is_chord": False,
                                "obj": el
                            }
                            merged_notes.append(entry)
                            merged_objs.append(el)
                            self.all_notes.append(entry["name"])
                            measure_has_notes = True

                        # CHORD
                        elif isinstance(el, music21.chord.Chord):
                            dur = float(el.quarterLength)
                            for p in el.pitches:
                                entry = {
                                    "name": self.normalize_note_name(p.name),
                                    "duration": dur,
                                    "is_chord": True,
                                    "obj": el
                                }
                                merged_notes.append(entry)
                                self.all_notes.append(entry["name"])
                            merged_objs.append(el)
                            measure_has_notes = True

            # If completely empty bar, add placeholder
            if not measure_has_notes:
                merged_notes.append({
                    "name": None,
                    "duration": 1.0,
                    "is_chord": False,
                    "obj": None
                })

            self.bars_notes.append(merged_notes)
            self.bars_notes_objs.append(merged_objs)

    def normalize_key_to_db(self, tonic_name, mode):
        """
        Convert tonic+mode (major/minor) into one of the keys in the database.
        """
        target = tonic_name + ("m" if mode == "minor" else "")

        # Direct hit
        if target in DATABASE["CHORDS_BY_KEY"]:
            return target

        # Fallback: use enharmonic equivalents
        try:
            pitch = music21.pitch.Pitch(tonic_name)
            enh = pitch.getEnharmonic().name  # get alternate spelling
            target2 = enh + ("m" if mode == "minor" else "")

            if target2 in DATABASE["CHORDS_BY_KEY"]:
                return target2
        except:
            pass

        # If still nothing, try scanning all DB keys with same pitch class
        for k in DATABASE["CHORDS_BY_KEY"].keys():
            k_tonic = k.replace("m", "")
            try:
                if music21.pitch.Pitch(k_tonic).pitchClass == music21.pitch.Pitch(tonic_name).pitchClass:
                    return k
            except:
                continue

        # worst case: return original tonic with mode
        return target

    def analyze_key_signature(self):
        """
        Melody-first key detection with MusicXML key signature check first.
        """
        # 1. CHECK MUSICXML KEY SIGNATURE FIRST
        ks = None
        try:
            if hasattr(self, 'score') and self.score.parts:
                part = self.score.parts[0]
                key_sigs = part.recurse().getElementsByClass(music21.key.KeySignature)
                if key_sigs:
                    ks = key_sigs[0]
        except Exception as e:
            print(f"Error checking key signature: {e}")
            ks = None

        # This check should be outside the try-except
        if ks is not None:
            sf = ks.sharps
            try:
                key_obj = music21.key.KeySignature(sf).asKey()
                tonic = key_obj.tonic.name
                mode = key_obj.mode
                final_key = self.normalize_key_to_db(tonic, mode)
                print(f"[KEY SIGNATURE] Found signature {sf} sharps → returning {final_key}")
                self.detected_key = final_key
                return final_key
            except Exception as e:
                print(f"Error processing key signature {sf}: {e}")
                # Continue with fallback detection

        # Fallback detection if no valid key signature found
        if not self.bars_notes_objs or not self.all_notes:
            self.detected_key = None
            return None

        # Counters (weighted)
        melodic_counter = Counter()   # top notes each measure (strong)
        bass_counter = Counter()      # bass notes each measure (medium)
        chord_tone_counter = Counter()# chord tones but lower weight
        raw_note_counter = Counter([self.normalize_note_name(n) for n in self.all_notes])

        # iterate measures and collect melody (top), bass (lowest), chord tone occurrences
        for measure_objs in self.bars_notes_objs:
            # collect pitched note entries with duration + midi
            measure_notes = []
            for el in measure_objs:
                if el is None:  # Skip None objects from empty bars
                    continue

                try:
                    if isinstance(el, music21.note.Rest):
                        continue
                    if isinstance(el, music21.note.Note):
                        name = self.normalize_note_name(el.pitch.name)
                        dur = float(getattr(el.duration, "quarterLength", 1.0) or 1.0)
                        midi = getattr(el.pitch, "midi", None)
                        measure_notes.append({'type': 'note', 'name': name, 'dur': dur, 'midi': midi, 'obj': el})
                    elif isinstance(el, music21.chord.Chord):
                        dur = float(getattr(el.duration, "quarterLength", 1.0) or 1.0)
                        # for chord pitches add entries but keep type chord
                        for p in el.pitches:
                            name = self.normalize_note_name(p.name)
                            try:
                                midi = p.midi
                            except Exception:
                                midi = None
                            measure_notes.append({'type': 'chord_tone', 'name': name, 'dur': dur, 'midi': midi, 'obj': el})
                    elif isinstance(el, music21.harmony.ChordSymbol):
                        # chord symbols don't contribute melodic votes directly here
                        continue
                except Exception as e:
                    print(f"Error processing element in measure: {e}")
                    continue

            if not measure_notes:
                continue

            # choose melodic (top) note: highest midi among notes with dur >= 0.25 if possible
            melodic_candidates = [n for n in measure_notes if n['dur'] >= 0.25]
            if not melodic_candidates:
                melodic_candidates = measure_notes  # fallback

            top = None
            top_midi = None
            for n in melodic_candidates:
                if n['midi'] is not None:
                    if top is None or n['midi'] > top_midi:
                        top = n
                        top_midi = n['midi']
            if top is None:
                # fallback: last candidate
                top = melodic_candidates[-1] if melodic_candidates else None

            if top:
                melodic_counter[top['name']] += 3   # strong weight for melody

            # choose bass: lowest midi among notes (prefer notes with dur>=0.25)
            bass_candidates = [n for n in measure_notes if n['midi'] is not None]
            if bass_candidates:
                lowest = min(bass_candidates, key=lambda x: x['midi'])
                bass_counter[lowest['name']] += 2

            # chord-tone collection: mark chord_tone entries but reduce their influence later
            for n in measure_notes:
                if n['type'] == 'chord_tone':
                    # ignore very short chord accidentals
                    if n['dur'] < 0.25:
                        continue
                    chord_tone_counter[n['name']] += 1

        # Compose an overall weighted counter for scoring keys
        combined_counter = Counter()
        for name, cnt in melodic_counter.items():
            combined_counter[name] += cnt * 1.0   # already strong value

        for name, cnt in bass_counter.items():
            combined_counter[name] += cnt * 0.8

        for name, cnt in chord_tone_counter.items():
            combined_counter[name] += cnt * 0.35

        for name, cnt in raw_note_counter.items():
            # include raw counts but low weight so short passing tones don't dominate
            combined_counter[name] += cnt * 0.25

        # Last note (cadence) normalized
        last_note_raw = self.all_notes[-1] if self.all_notes else None
        last_note = self.normalize_note_name(last_note_raw) if last_note_raw else None

        CADENCE_BONUS = 40

        best_key = None
        best_score = -1e9

        # Evaluate each major and its relative natural minor (constructed)
        for maj in DATABASE["MAJOR_KEYS"]:
            maj_scale = DATABASE["MAJOR_NOTES"].get(maj, [])
            if not maj_scale:
                continue

            norm_maj_scale = [self.normalize_note_name(n) for n in maj_scale]

            # major score from combined_counter
            maj_score = sum(combined_counter.get(n, 0) for n in norm_maj_scale)

            # relative natural minor built by rotation (6th degree)
            rel_nat_minor = maj_scale[5:] + maj_scale[:5]
            norm_rel_minor = [self.normalize_note_name(n) for n in rel_nat_minor]
            rel_score = sum(combined_counter.get(n, 0) for n in norm_rel_minor)

            # cadence bonus: add when last note equals tonic
            if last_note:
                if last_note == self.normalize_note_name(maj):
                    maj_score += CADENCE_BONUS
                rel_tonic = norm_rel_minor[0] if norm_rel_minor else None
                if rel_tonic and last_note == rel_tonic:
                    rel_score += CADENCE_BONUS

            # small tie-breaker bonus: if dominant (5th) appears in melodic_counter, add a bit
            maj_fifth = norm_maj_scale[4] if len(norm_maj_scale) > 4 else None
            if maj_fifth and melodic_counter.get(maj_fifth, 0) > 0:
                maj_score += 2

            # pick best between the major and the relative minor and compare to global best
            if maj_score > best_score:
                best_score = maj_score
                best_key = maj
            if rel_score > best_score:
                best_score = rel_score
                best_key = DATABASE["MINOR_KEYS"].get(maj, None)

        print(f"(melody-first) detected key: {best_key} (score {best_score})")
        self.detected_key = best_key
        return best_key

    def check_major_minor(self, key, notes_in_song=None):
        """Enhanced minor key detection with multiple indicators"""
        if notes_in_song is None:
            notes_in_song = self.all_notes

        if not key or key.endswith("m"):
            return key

        if key not in DATABASE["MINOR_KEYS"]:
            return key

        norm_song = {self.normalize_note_name(n) for n in notes_in_song}
        maj_scale = DATABASE["MAJOR_NOTES"].get(key, [])

        if not maj_scale:
            return key

        # Check multiple minor indicators
        minor_indicators = 0

        # 1. Check raised 7th (harmonic minor)
        seventh = self.normalize_note_name(maj_scale[6]) if len(maj_scale) > 6 else None
        raised7 = self.half_step_up(seventh) if seventh else None
        if raised7 and self.normalize_note_name(raised7) in norm_song:
            minor_indicators += 2

        # 2. Check raised 6th (melodic minor)
        sixth = self.normalize_note_name(maj_scale[5]) if len(maj_scale) > 5 else None
        raised6 = self.half_step_up(sixth) if sixth else None
        if raised6 and self.normalize_note_name(raised6) in norm_song:
            minor_indicators += 1

        # 3. Check raised 5th (already in your code)
        fifth = self.normalize_note_name(maj_scale[4]) if len(maj_scale) > 4 else None
        raised5 = self.half_step_up(fifth) if fifth else None
        if raised5 and self.normalize_note_name(raised5) in norm_song:
            minor_indicators += 2

        # 4. Check if minor tonic appears more than major tonic
        minor_tonic = DATABASE["MINOR_KEYS"][key]
        minor_tonic_norm = self.normalize_note_name(minor_tonic.replace("m", ""))
        major_tonic_norm = self.normalize_note_name(key)

        major_count = sum(1 for n in notes_in_song if self.normalize_note_name(n) == major_tonic_norm)
        minor_count = sum(1 for n in notes_in_song if self.normalize_note_name(n) == minor_tonic_norm)

        if minor_count > major_count:
            minor_indicators += 1

        # Switch to minor if strong indicators present
        if minor_indicators >= 3:
            return DATABASE["MINOR_KEYS"][key]

        return key

    def half_step_up(self, note):
        """Helper for half-step up"""
        if not note:
            return None

        half_step_map = {
            "C": "C#", "C#": "D", "D": "D#", "D#": "E", "E": "F",
            "F": "F#", "F#": "G", "G": "G#", "G#": "A", "A": "A#", "A#": "B", "B": "C"
        }
        return half_step_map.get(note, note)

    def get_chord_candidates(self, key):
        chord_candidates = {}

        if not key:
            return {}

        # Get chord progression for the key
        if key in DATABASE["CHORDS_BY_KEY"]:
            chords = DATABASE["CHORDS_BY_KEY"][key]
        else:
            # Try relative major for minor keys
            rel_major = None
            for maj, minv in DATABASE["MINOR_KEYS"].items():
                if minv == key:
                    rel_major = maj
                    break
            chords = DATABASE["CHORDS_BY_KEY"].get(rel_major, [])

        # Build chord candidates with notes
        for chord in chords:
            if chord in DATABASE["CHORD_NOTES"]:
                chord_candidates[chord] = DATABASE["CHORD_NOTES"][chord]
            else:
                # Try to construct chord if missing (implement these helper methods)
                notes = self.construct_chord(chord)
                if notes:
                    chord_candidates[chord] = notes

        return chord_candidates

    def construct_chord(self, chord_name):
        """Construct chord notes for chords missing from database"""
        if chord_name.endswith('dim'):
            base = chord_name[:-3]
            # Construct diminished chord: root, minor third, diminished fifth
            root = self.normalize_note_name(base)
            third = self.half_step_up(self.half_step_up(root))  # Minor third = 3 half steps
            fifth = self.half_step_up(self.half_step_up(third))  # Diminished fifth = 3 half steps from third
            return [root, third, fifth] if all([root, third, fifth]) else None
        elif chord_name.endswith('m'):
            base = chord_name[:-1]
            # Construct minor chord: root, minor third, perfect fifth
            root = self.normalize_note_name(base)
            third = self.half_step_up(self.half_step_up(root))  # Minor third = 3 half steps
            fifth = self.half_step_up(self.half_step_up(self.half_step_up(third)))  # Perfect fifth = 4 half steps from third
            return [root, third, fifth] if all([root, third, fifth]) else None
        else:
            # Construct major chord: root, major third, perfect fifth
            root = self.normalize_note_name(chord_name)
            third = self.half_step_up(self.half_step_up(self.half_step_up(root)))  # Major third = 4 half steps
            fifth = self.half_step_up(self.half_step_up(third))  # Perfect fifth = 3 half steps from third
            return [root, third, fifth] if all([root, third, fifth]) else None

    def chord_degree(self, chord_name, key_name):
        if not key_name or not chord_name:
            return None

        if key_name in DATABASE["CHORDS_BY_KEY"]:
            scale = DATABASE["CHORDS_BY_KEY"][key_name]
        else:
            rel_major = None
            for maj, minv in DATABASE["MINOR_KEYS"].items():
                if minv == key_name:
                    rel_major = maj
                    break
            if rel_major and rel_major in DATABASE["CHORDS_BY_KEY"]:
                scale = DATABASE["CHORDS_BY_KEY"][rel_major]
            else:
                scale = DATABASE["CHORDS_BY_KEY"].get(key_name, None)

        if not scale:
            return None

        try:
            return scale.index(chord_name) + 1
        except ValueError:
            # Try chord without extensions
            base_chord = chord_name
            if chord_name.endswith(('7', 'maj7', 'min7', 'dim7', 'aug')):
                # Remove common extensions
                base_chord = re.sub(r'(7|maj7|min7|dim7|aug)$', '', chord_name)

            try:
                return scale.index(base_chord) + 1
            except ValueError:
                return None

    def get_bass_of_bar(self, bar_objs):
        if not bar_objs:
            return None
        lowest = None
        for el in bar_objs:
            if el is None:
                continue
            try:
                if isinstance(el, music21.chord.Chord):
                    for p in el.pitches:
                        if lowest is None or p.midi < lowest.midi:
                            lowest = p
                elif isinstance(el, music21.note.Note):
                    p = el.pitch
                    if lowest is None or p.midi < lowest.midi:
                        lowest = p
            except Exception:
                continue
        return self.normalize_note_name(lowest.name) if lowest else None

    def find_best_chord_for_bar(self, bar_note_names, bar_objs, chord_candidates, prev_chord=None, bar_index=0, total_bars=1):
        """
        Improved chord selection with configurable weights.
        """
        print(f"\n================ BAR {bar_index+1} ================")
        print("Notes:", bar_note_names)
        print("Prev chord:", prev_chord)
        print("-----------------------------------------------")

        # Filter out None notes
        bar_note_names = [n for n in bar_note_names if n is not None]

        try:
            if bar_objs:
                for el in bar_objs:
                    if el is None:
                        continue
                    # music21 chord symbol (common when chords are written as symbols)
                    if isinstance(el, music21.harmony.ChordSymbol):
                        # figure e.g. 'F', 'Am', etc.
                        figure = getattr(el, 'figure', None)
                        if figure:
                            norm_fig = figure.strip()
                            print(f"Manual chord symbol found in bar: {norm_fig} -> using it.")
                            return norm_fig

                    # custom user annotation (if you attach 'manual_chord' attribute to an element)
                    if hasattr(el, 'manual_chord'):
                        mc = getattr(el, 'manual_chord')
                        if mc:
                            print(f"Manual chord annotation found in bar: {mc} -> using it.")
                            return mc
        except Exception:
            pass

        # Build a list of note occurrences with rough durations and midi for weighting.
        note_occurrences = []   # list of dicts: {'name':str, 'dur':float, 'midi':int}
        try:
            for el in (bar_objs or []):
                if el is None:
                    continue
                if isinstance(el, music21.note.Rest):
                    continue
                if isinstance(el, music21.note.Note):
                    name = self.normalize_note_name(el.pitch.name)
                    dur = getattr(el.duration, "quarterLength", 1.0) or 1.0
                    midi = getattr(el.pitch, "midi", None)
                    note_occurrences.append({'name': name, 'dur': float(dur), 'midi': midi})
                elif isinstance(el, music21.chord.Chord):
                    dur = getattr(el.duration, "quarterLength", 1.0) or 1.0
                    # for chords, add each pitch as separate occurrence
                    for p in el.pitches:
                        name = self.normalize_note_name(p.name)
                        try:
                            midi = p.midi
                        except Exception:
                            midi = None
                        note_occurrences.append({'name': name, 'dur': float(dur), 'midi': midi})
        except Exception:
            # fallback: use names only with default durations
            note_occurrences = [{'name': n, 'dur': 1.0, 'midi': None} for n in bar_note_names]

        # If bar_objs missing, infer from bar_note_names
        if not note_occurrences and bar_note_names:
            note_occurrences = [{'name': n, 'dur': 1.0, 'midi': None} for n in bar_note_names]

        total_note_dur = sum(n['dur'] for n in note_occurrences) or 1.0

        # Find top and bass note (by midi if available)
        top_note = None
        lowest_midi = None
        for n in note_occurrences:
            if n['midi'] is not None:
                if top_note is None or (n['midi'] > top_note['midi']):
                    top_note = n
                if lowest_midi is None or n['midi'] < lowest_midi:
                    lowest_midi = n['midi']
        # If midi not available, fallback to lexical top (not great but safe)
        if top_note is None and note_occurrences:
            # top as last note in the treble extraction order
            top_note = note_occurrences[-1]

        bass = self.get_bass_of_bar(bar_objs)
        print(f"(computed) bass from bar objects = {bass}")

        # 1) Base matching with configurable weights
        scores = {}
        max_base = -1e9
        print(">> BASE SCORE (duration-weighted tone matches)")

        for chord, chord_notes in chord_candidates.items():
            # chord_notes is expected as list like [root, third, fifth] (may be missing)
            root = chord_notes[0] if len(chord_notes) > 0 else None
            third = chord_notes[1] if len(chord_notes) > 1 else None
            fifth = chord_notes[2] if len(chord_notes) > 2 else None
            extensions = chord_notes[3:] if len(chord_notes) > 3 else []

            base = 0.0
            # Use configurable weights
            tone_weights = {
                root: self.weights['root_in_bass'] if root else -self.weights['short_note_penalty'],
                third: self.weights['third_in_bass'] if third else -self.weights['short_note_penalty'],
                fifth: self.weights['fifth_in_bass'] if fifth else -self.weights['short_note_penalty']
            }

            # small weight for extensions
            ext_weight = 0.8 * self.weights['base_match']

            for occ in note_occurrences:
                name = occ['name']
                dur = occ['dur']
                dur_factor = min(dur / total_note_dur * 4.0, 2.0) * self.weights['duration_factor']
                # reduce influence of very short notes (likely passing tones)
                if dur < 0.25:
                    dur_factor *= 0.45

                if name in tone_weights and tone_weights[name] > 0:
                    base += tone_weights[name] * dur_factor
                elif name in extensions:
                    base += ext_weight * dur_factor
                else:
                    # small penalty for extraneous notes
                    base -= self.weights['short_note_penalty'] * dur_factor

            # penalize if chord has zero intersection with bar notes
            intersection = set([n['name'] for n in note_occurrences]) & set([root, third, fifth] + extensions)
            if not intersection:
                base -= 2.0 * self.weights['base_match']

            scores[chord] = {
                'base': base, 'root': root, 'third': third, 'fifth': fifth,
                'extensions': extensions
            }
            print(f"  {chord:<8} root={str(root):<3} third={str(third):<3} fifth={str(fifth):<3} -> base={base:.3f}")
            if base > max_base:
                max_base = base

        # 2) Candidate set with tolerance
        tol = 1.0 * self.weights['base_match']
        candidates = [c for c, s in scores.items() if s['base'] >= max_base - tol]
        if not candidates:
            candidates = list(scores.keys())

        print("\n>> CHORDS IN TOLERANCE (max_base =", max_base, ")")
        for c in candidates:
            print("  ", c)

        # 3) strong-tone & bass scoring with configurable weights
        print("\n>> STRONG + BASS SCORE (bass =", bass, ")")
        for c in candidates:
            s = scores[c]
            st = 0.0

            # count weighted occurrences of root/third/fifth across note_occurrences
            for occ in note_occurrences:
                name = occ['name']
                dur = occ['dur']
                dur_factor = min(dur, 2.0) * self.weights['duration_factor']
                if name == s['root']:
                    st += self.weights['root_in_bass'] * dur_factor
                elif name == s['third']:
                    st += self.weights['third_in_bass'] * dur_factor
                elif name == s['fifth']:
                    st += self.weights['fifth_in_bass'] * dur_factor
                elif name in s.get('extensions', []):
                    st += 0.8 * self.weights['base_match'] * dur_factor

            # bass match bonus (strong cue)
            if bass and s['root'] and bass == s['root']:
                st += self.weights['bass_match']
            elif bass and s['fifth'] and bass == s['fifth']:
                st += self.weights['fifth_in_bass'] * 1.5
            elif bass and s['third'] and bass == s['third']:
                st += self.weights['third_in_bass'] * 1.2

            # top-note (melody) bonus
            top_bonus = 0.0
            if top_note:
                top_name = top_note['name']
                if top_name == s['root']:
                    top_bonus += self.weights['top_note_root']
                elif top_name == s['third']:
                    top_bonus += self.weights['top_note_third']
                elif top_name == s['fifth']:
                    top_bonus += self.weights['top_note_fifth']
                elif top_name in s.get('extensions', []):
                    top_bonus += 0.6 * self.weights['base_match']
            st += top_bonus

            scores[c]['strong'] = st
            print(f"  {c:<8} strong={st:.3f} (top_bonus={top_bonus:.2f})")

        # 4) progression/context scoring
        print("\n>> PROGRESSION SCORE")
        common_pairs = set()
        for prog in DATABASE.get("COMMON_CHORD_PROGRESSIONS", []):
            for i in range(len(prog) - 1):
                common_pairs.add((prog[i], prog[i+1]))

        for c in candidates:
            prog = 0.0
            if prev_chord:
                prev_deg = self.chord_degree(prev_chord, self.detected_key)
                curr_deg = self.chord_degree(c, self.detected_key)
                if prev_deg and curr_deg:
                    # Use configurable weights for cadences
                    if prev_deg == 5 and curr_deg == 1:
                        prog += self.weights['cadence_5_1']
                    if prev_deg == 4 and curr_deg == 1:
                        prog += self.weights['cadence_4_1']
                    if prev_deg == 5 and curr_deg == 6:
                        prog += 0.8 * self.weights['base_match']
                    if (prev_deg, curr_deg) in common_pairs:
                        prog += 2.0 * self.weights['base_match']
                    if abs(prev_deg - curr_deg) == 1:
                        prog += 0.6 * self.weights['base_match']
                    if abs(prev_deg - curr_deg) > 3:
                        prog -= 0.8 * self.weights['base_match']

            scores[c]['prog'] = prog
            print(f"  {c:<8} prog={prog:.3f}")

        # 5) special/cadential score (first/last bar etc.)
        print("\n>> SPECIAL (FIRST/LAST BAR / CADENCE)")

        for c in candidates:
            special = 0.0
            deg = self.chord_degree(c, self.detected_key)

            # Use configurable weights for first/last bar
            if bar_index == 0:
                if deg == 1:
                    special += self.weights['first_bar_tonic']
                elif deg in (3, 5):
                    special += 1.5 * self.weights['base_match']

            # last bar: strong tonic bias (cadence)
            if bar_index == total_bars - 1:
                if deg == 1:
                    special += self.weights['last_bar_tonic']
                elif deg == 5:
                    special += 4.0 * self.weights['base_match']

            # penalize highly unlikely degrees for cadential positions
            if bar_index == total_bars - 1 and deg in (2, 3):
                special -= 1.0 * self.weights['base_match']

            scores[c]['special'] = special
            print(f"  {c:<8} special={special:.3f}")

        # 6) Non-harmonic / passing tone penalty
        short_notes = sum(1 for occ in note_occurrences if occ['dur'] < 0.25)
        short_penalty = short_notes * self.weights['short_note_penalty']
        if short_notes:
            print(f"\n>> NON-HARMONIC INFO: short_notes={short_notes} penalty={short_penalty:.2f}")
        else:
            print("\n>> NON-HARMONIC INFO: none detected")

        # 7) total score
        print("\n>> TOTAL SCORE")
        best = None
        best_total = -1e9
        for c in candidates:
            total = scores[c]['base'] + scores[c]['strong'] + scores[c]['prog'] + scores[c]['special']

            # subtract penalty if chord relies mainly on short notes
            long_match_weight = 0.0
            for occ in note_occurrences:
                if occ['dur'] >= 0.25:
                    name = occ['name']
                    if name in [scores[c]['root'], scores[c]['third'], scores[c]['fifth']] or name in scores[c].get('extensions', []):
                        long_match_weight += min(occ['dur'], 2.0)

            if long_match_weight < 0.5 and scores[c]['base'] > 0:
                total -= 1.5 * self.weights['base_match']

            if short_notes:
                total -= short_penalty * (1.0 if long_match_weight < 1.0 else 0.2)

            scores[c]['total'] = total
            print(f"  {c:<8} total={total:.3f} (base={scores[c]['base']:.3f} strong={scores[c]['strong']:.3f} prog={scores[c]['prog']:.3f} special={scores[c]['special']:.3f})")

            if total > best_total:
                best_total = total
                best = c

        # 8) tie-breaker
        ties = [c for c in candidates if abs(scores[c]['total'] - best_total) < 1e-9]
        if len(ties) > 1:
            ties.sort(key=lambda x: (
                # prefer exact bass root match
                1 if (bass and scores[x]['root'] and bass == scores[x]['root']) else 0,
                scores[x]['strong'],
                scores[x]['base']
            ), reverse=True)
            best = ties[0]

        print("\n>>> SELECTED CHORD:", best)
        print("===================================================")
        return best

    def analyze_chord_progression(self):
        print("=== MUSIC ANALYSIS ===")

        # Step 1: detect key
        detected_key = self.analyze_key_signature()
        print(f"Detected key from analysis: {detected_key}")

        if self.key_signature and self.key_signature != "None" and self.key_signature in DATABASE["MAJOR_KEYS"]:
            final_key = self.key_signature
            print(f"Using key signature from file: {final_key}")
        else:
            final_key = detected_key
            print(f"Using analyzed key: {final_key}")

        # check major/minor
        final_key = self.check_major_minor(final_key, self.all_notes)
        self.detected_key = final_key  # ensure stored

        # chord candidates (dict chord -> notes)
        chord_candidates = self.get_chord_candidates(final_key)
        print(f"Chord candidates for {final_key}: {list(chord_candidates.keys())}")

        # analyze per bar
        chord_progression = []
        prev_chord = None
        total_bars = len(self.bars_notes)

        print("\n=== BAR-BY-BAR ANALYSIS ===")
        for i, (bar_names, bar_objs) in enumerate(zip(self.bars_notes, self.bars_notes_objs)):
            # Extract note names from bar entries
            bar_note_list = [entry["name"] for entry in bar_names if entry["name"] is not None]
            best = self.find_best_chord_for_bar(bar_note_list, bar_objs, chord_candidates, prev_chord=prev_chord, bar_index=i, total_bars=total_bars)
            chord_progression.append(best)
            prev_chord = best

            chord_notes = chord_candidates.get(best, [])
            match_count = sum(1 for note in bar_note_list if note in chord_notes)
            print(f"Bar {i+1}: Notes {bar_note_list}")
            print(f"         -> Chord: {best} {chord_notes} (match: {match_count}/{len(bar_note_list)} notes)")

        # Roman numerals
        if final_key in DATABASE["CHORDS_BY_KEY"]:
            chords_in_key = DATABASE["CHORDS_BY_KEY"][final_key]
        else:
            relative_major = None
            for major, minor in DATABASE["MINOR_KEYS"].items():
                if minor == final_key:
                    relative_major = major
                    break
            chords_in_key = DATABASE["CHORDS_BY_KEY"].get(relative_major, DATABASE["CHORDS_BY_KEY"]["C"])

        roman_numerals = []
        for ch in chord_progression:
            if ch in chords_in_key:
                pos = chords_in_key.index(ch) + 1
                roman = {1: "I", 2: "ii", 3: "iii", 4: "IV", 5: "V", 6: "vi", 7: "vii°"}.get(pos, "?")
                roman_numerals.append(roman)
            else:
                roman_numerals.append("?")

        print(f"\n=== FINAL RESULTS ===")
        print(f"Key: {final_key}")
        print(f"Chord Progression: {' - '.join(chord_progression)}")
        print(f"Roman Numerals: {' - '.join(roman_numerals)}")

        return chord_progression
```

```python

class MusicAnalyzer:
    def __init__(self, musicxml_file):
        self.musicxml_file = musicxml_file
        self.score = None
        self.key_signature = None
        self.detected_key = None
        self.bars_notes = []
        self.bars_notes_objs = []
        self.all_notes = []

        self.load_musicxml()

    # Loading & extraction
    def load_musicxml(self):
        try:
            print(f"Loading MusicXML file: {self.musicxml_file}")
            self.score = music21.converter.parse(self.musicxml_file)

            # Ambil key dari file
            key_elements = self.score.flat.getElementsByClass(music21.key.Key)
            if key_elements:
                self.key_signature = str(key_elements[0])
            else:
                key_sig_elements = self.score.flat.getElementsByClass(music21.key.KeySignature)
                if key_sig_elements:
                    self.key_signature = str(key_sig_elements[0])
                else:
                    self.key_signature = None

            self.extract_notes()

            print(f"Successfully loaded file. Found {len(self.bars_notes)} bars.")
        except Exception as e:
            print(f"Error loading MusicXML file: {e}")
            raise

    def normalize_note_name(self, note_name):
        if not note_name:
            return None

        # Remove octave numbers
        name = re.sub(r'\d', '', note_name)

        # Normalize unicode accidentals
        name = name.replace("♯", "#").replace("♭", "b").replace("♮", "")

        # Master enharmonic conversion table (flats → sharps)
        ENH = {
            "Cb": "B",
            "B#": "C",
            "Db": "C#",
            "D#": "D#",
            "Eb": "D#",
            "E#": "F",
            "Fb": "E",
            "Gb": "F#",
            "G#": "G#",
            "Ab": "G#",
            "A#": "A#",
            "Bb": "A#",
        }

        # If already natural
        if name in ["A", "B", "C", "D", "E", "F", "G"]:
            return name

        # Convert single accidental notes
        if name in ENH:
            return ENH[name]

        # Edge-case: unexpected patterns → try stripping unknown chars
        base = re.sub(r'[^A-G#b]', '', name)
        return ENH.get(base, base)


    def extract_notes(self):
        """
        Extract notes from all parts and store duration + pitch + type.
        Each bar will contain entries like:
            {
                "name": "C",
                "duration": 1.0,
                "is_chord": False,
                "obj": <music21.note.Note>
            }
        """
        self.all_notes = []
        self.bars_notes = []
        self.bars_notes_objs = []

        if not self.score.parts:
            return

        parts_measures = []
        max_measure_number = 0

        # Collect all measure
        for part in self.score.parts:
            pm = list(part.getElementsByClass(music21.stream.Measure))
            parts_measures.append(pm)
            max_measure_number = max(max_measure_number, len(pm))

        # For each measure
        for i in range(max_measure_number):

            merged_notes = []
            merged_objs = []

            for part_measure_list in parts_measures:
                if i < len(part_measure_list):
                    measure = part_measure_list[i]

                    for el in measure.notesAndRests:

                        # Skip rests
                        if isinstance(el, music21.note.Rest):
                            continue

                        # Single NOTE
                        if isinstance(el, music21.note.Note):
                            entry = {
                                "name": self.normalize_note_name(el.pitch.name),
                                "duration": float(el.quarterLength),
                                "is_chord": False,
                                "obj": el
                            }
                            merged_notes.append(entry)
                            merged_objs.append(el)
                            self.all_notes.append(entry["name"])

                        # CHORD
                        elif isinstance(el, music21.chord.Chord):
                            dur = float(el.quarterLength)
                            for p in el.pitches:
                                entry = {
                                    "name": self.normalize_note_name(p.name),
                                    "duration": dur,
                                    "is_chord": True,
                                    "obj": el
                                }
                                merged_notes.append(entry)
                                self.all_notes.append(entry["name"])
                            merged_objs.append(el)

            # Assign
            self.bars_notes.append(merged_notes)
            self.bars_notes_objs.append(merged_objs)


    # def analyze_key_signature(self):
    #     """
    #     Simple, robust signature/key detection:
    #     - Count occurrences of notes that belong to each MAJOR scale
    #     - For each major also compute relative-natural-minor scale score (shifted major)
    #     - Add a strong cadence bonus if the last note equals tonic of major or relative minor
    #     - Choose the best-scoring key (major or its relative minor)
    #     """
    #     if not self.all_notes:
    #         self.detected_key = None
    #         return None
    #
    #     # normalize all song notes once
    #     norm_song = [self.normalize_note_name(n) for n in self.all_notes]
    #     note_counter = Counter(norm_song)
    #
    #     # normalized last note (cadence)
    #     last_note = self.normalize_note_name(self.all_notes[-1]) if self.all_notes else None
    #
    #     best_key = None
    #     best_score = -1e9
    #
    #     # cadence bonus (tunable)
    #     CADENCE_BONUS = 80  # large but not insane; makes final note decisive when present
    #
    #     for maj in DATABASE["MAJOR_KEYS"]:
    #         maj_scale = DATABASE["MAJOR_NOTES"].get(maj, [])
    #         norm_maj_scale = [self.normalize_note_name(n) for n in maj_scale]
    #
    #         # major score = sum of occurrences of major scale tones
    #         maj_score = sum(note_counter.get(n, 0) for n in norm_maj_scale)
    #
    #         # construct natural minor from major by rotating (relative minor = 6th degree)
    #         # natural minor scale = majorscale[5:] + majorscale[:5]
    #         rel_nat_minor = maj_scale[5:] + maj_scale[:5]
    #         norm_rel_minor = [self.normalize_note_name(n) for n in rel_nat_minor]
    #
    #         rel_score = sum(note_counter.get(n, 0) for n in norm_rel_minor)
    #
    #         # cadence bonus: if last note equals tonic of major or relative minor
    #         if last_note:
    #             if last_note == self.normalize_note_name(maj):
    #                 maj_score += CADENCE_BONUS
    #             # rel_minor tonic is the 6th degree of major
    #             rel_tonic = norm_rel_minor[0] if norm_rel_minor else None
    #             if rel_tonic and last_note == rel_tonic:
    #                 rel_score += CADENCE_BONUS
    #
    #         # Compare which of the pair (maj or its relative minor) is better
    #         if rel_score > best_score:
    #             best_score = rel_score
    #             # get minor key name from DB
    #             best_key = DATABASE["MINOR_KEYS"].get(maj, None)
    #         if maj_score > best_score:
    #             best_score = maj_score
    #             best_key = maj
    #
    #     # finalize
    #     print(f"(fallback) detected key: {best_key} (score {best_score})")
    #     self.detected_key = best_key
    #     return best_key
    def normalize_key_to_db(self, tonic_name, mode):
        """
        Convert tonic+mode (major/minor) into one of the keys in the database.
        Example:
            tonic='Bb', mode='minor' → 'Bbm'
            tonic='A#', mode='minor' → 'A#m'
            tonic='Cb', mode='major' → 'Cb'
        """
        target = tonic_name + ("m" if mode == "minor" else "")

        # Direct hit
        if target in DATABASE["CHORDS_BY_KEY"]:
            return target

        # Fallback: use enharmonic equivalents
        try:
            pitch = music21.pitch.Pitch(tonic_name)
            enh = pitch.getEnharmonic().name  # get alternate spelling
            target2 = enh + ("m" if mode == "minor" else "")

            if target2 in DATABASE["CHORDS_BY_KEY"]:
                return target2
        except:
            pass

        # If still nothing, try scanning all DB keys with same pitch class
        for k in DATABASE["CHORDS_BY_KEY"].keys():
            k_tonic = k.replace("m", "")
            try:
                if music21.pitch.Pitch(k_tonic).pitchClass == music21.pitch.Pitch(tonic_name).pitchClass:
                    return k
            except:
                continue

        # worst case: return original tonic with mode
        return target

    def analyze_key_signature(self):
        """
        Melody-first key detection:
          - weight melodic (top) notes strongly,
          - weight bass notes moderately,
          - weight chord tones lightly,
          - ignore very short passing tones (dur < 0.25) for key votes,
          - add cadence (last note) bonus.
        Returns a major key or its relative minor (string).
        """

        # 1. CHECK MUSICXML KEY SIGNATURE FIRST
# 1. CHECK MUSICXML KEY SIGNATURE FIRST
        ks = None
        try:
            part = self.score.parts[0] if hasattr(self, 'score') else None
            if part:
                key_sigs = part.recurse().getElementsByClass(music21.key.KeySignature)
                if key_sigs:
                    ks = key_sigs[0]     # take the first signature
        except:
            ks = None

        # 🔥 FIX: Check outside the try/except so it ALWAYS runs if ks was found
        if ks:
            sf = ks.sharps
            key_obj = music21.key.KeySignature(sf).asKey()

            tonic = key_obj.tonic.name
            mode = key_obj.mode

            final_key = self.normalize_key_to_db(tonic, mode)

            print(f"[KEY SIGNATURE] Found signature {sf} → returning {final_key}")
            self.detected_key = final_key
            return final_key






        if not self.bars_notes_objs or not self.all_notes:
            self.detected_key = None
            return None

        # Counters (weighted)
        melodic_counter = Counter()   # top notes each measure (strong)
        bass_counter = Counter()      # bass notes each measure (medium)
        chord_tone_counter = Counter()# chord tones but lower weight
        raw_note_counter = Counter([self.normalize_note_name(n) for n in self.all_notes])

        # iterate measures and collect melody (top), bass (lowest), chord tone occurrences
        for measure_objs in self.bars_notes_objs:
            # collect pitched note entries with duration + midi
            measure_notes = []
            for el in measure_objs:
                try:
                    if isinstance(el, music21.note.Rest):
                        continue
                    if isinstance(el, music21.note.Note):
                        name = self.normalize_note_name(el.pitch.name)
                        dur = float(getattr(el.duration, "quarterLength", 1.0) or 1.0)
                        midi = getattr(el.pitch, "midi", None)
                        measure_notes.append({'type': 'note', 'name': name, 'dur': dur, 'midi': midi, 'obj': el})
                    elif isinstance(el, music21.chord.Chord):
                        dur = float(getattr(el.duration, "quarterLength", 1.0) or 1.0)
                        # for chord pitches add entries but keep type chord
                        for p in el.pitches:
                            name = self.normalize_note_name(p.name)
                            try:
                                midi = p.midi
                            except Exception:
                                midi = None
                            measure_notes.append({'type': 'chord_tone', 'name': name, 'dur': dur, 'midi': midi, 'obj': el})
                    elif isinstance(el, music21.harmony.ChordSymbol):
                        # chord symbols don't contribute melodic votes directly here
                        continue
                except Exception:
                    continue

            if not measure_notes:
                continue

            # choose melodic (top) note: highest midi among notes with dur >= 0.25 if possible
            melodic_candidates = [n for n in measure_notes if n['dur'] >= 0.25]
            if not melodic_candidates:
                melodic_candidates = measure_notes  # fallback

            top = None
            top_midi = None
            for n in melodic_candidates:
                if n['midi'] is not None:
                    if top is None or n['midi'] > top_midi:
                        top = n
                        top_midi = n['midi']
            if top is None:
                # fallback: last candidate
                top = melodic_candidates[-1]

            melodic_counter[top['name']] += 3   # strong weight for melody

            # choose bass: lowest midi among notes (prefer notes with dur>=0.25)
            bass_candidates = [n for n in measure_notes if n['midi'] is not None]
            if bass_candidates:
                lowest = min(bass_candidates, key=lambda x: x['midi'])
                bass_counter[lowest['name']] += 2

            # chord-tone collection: mark chord_tone entries but reduce their influence later
            for n in measure_notes:
                if n['type'] == 'chord_tone':
                    # ignore very short chord accidentals
                    if n['dur'] < 0.25:
                        continue
                    chord_tone_counter[n['name']] += 1

        # Compose an overall weighted counter for scoring keys
        # weights: melodic 3, bass 2, chord_tone 0.4, raw_occurrence 0.6 (to capture important monophonic notes)
        combined_counter = Counter()
        for name, cnt in melodic_counter.items():
            combined_counter[name] += cnt * 1.0   # already strong value

        for name, cnt in bass_counter.items():
            combined_counter[name] += cnt * 0.8

        for name, cnt in chord_tone_counter.items():
            combined_counter[name] += cnt * 0.35

        for name, cnt in raw_note_counter.items():
            # include raw counts but low weight so short passing tones don't dominate
            combined_counter[name] += cnt * 0.25

        # Last note (cadence) normalized
        last_note_raw = self.all_notes[-1] if self.all_notes else None
        last_note = self.normalize_note_name(last_note_raw) if last_note_raw else None

        CADENCE_BONUS = 40

        best_key = None
        best_score = -1e9

        # Evaluate each major and its relative natural minor (constructed)
        for maj in DATABASE["MAJOR_KEYS"]:
            maj_scale = DATABASE["MAJOR_NOTES"].get(maj, [])
            norm_maj_scale = [self.normalize_note_name(n) for n in maj_scale]

            # major score from combined_counter
            maj_score = sum(combined_counter.get(n, 0) for n in norm_maj_scale)

            # relative natural minor built by rotation (6th degree)
            rel_nat_minor = maj_scale[5:] + maj_scale[:5]
            norm_rel_minor = [self.normalize_note_name(n) for n in rel_nat_minor]
            rel_score = sum(combined_counter.get(n, 0) for n in norm_rel_minor)

            # cadence bonus: add when last note equals tonic
            if last_note:
                if last_note == self.normalize_note_name(maj):
                    maj_score += CADENCE_BONUS
                rel_tonic = norm_rel_minor[0] if norm_rel_minor else None
                if rel_tonic and last_note == rel_tonic:
                    rel_score += CADENCE_BONUS

            # small tie-breaker bonus: if dominant (5th) appears in melodic_counter, add a bit
            maj_fifth = norm_maj_scale[4] if len(norm_maj_scale) > 4 else None
            if maj_fifth and melodic_counter.get(maj_fifth, 0) > 0:
                maj_score += 2

            # normalize accidentals: if a pitch appears only as very short notes across the piece, don't count it
            # (we already filtered short chord accidentals earlier)

            # pick best between the major and the relative minor and compare to global best
            if maj_score > best_score:
                best_score = maj_score
                best_key = maj
            if rel_score > best_score:
                best_score = rel_score
                best_key = DATABASE["MINOR_KEYS"].get(maj, None)

        print(f"(melody-first) detected key: {best_key} (score {best_score})")
        self.detected_key = best_key
        return best_key

    def check_major_minor(self, key, notes_in_song=None):
        """
        Simple post-filter:
        - If the raised 5th of the major key appears in the song => choose relative minor
          (user requested heuristic: raise the 5th of the major and check presence)
        - Otherwise remain in the provided key.
        """
        if notes_in_song is None:
            notes_in_song = self.all_notes

        if not key:
            return key

        # already minor — nothing to do
        if key.endswith("m"):
            return key

        # ensure key maps to a relative minor
        if key not in DATABASE["MINOR_KEYS"]:
            return key

        norm_song = {self.normalize_note_name(n) for n in notes_in_song}

        # major scale and its 5th
        maj_scale = DATABASE["MAJOR_NOTES"].get(key, [])
        if not maj_scale:
            return key

        fifth = self.normalize_note_name(maj_scale[4]) if len(maj_scale) > 4 else None

        # raise a half-step helper
        half_step_up = {
            "C":"C#","C#":"D","Db":"D","D":"D#","D#":"E","Eb":"E",
            "E":"F","Fb":"F","E#":"F#","F":"F#","F#":"G","Gb":"G",
            "G":"G#","G#":"A","Ab":"A","A":"A#","A#":"B","Bb":"B",
            "B":"C","Cb":"C","B#":"C#"
        }

        raised5 = half_step_up.get(fifth, None) if fifth else None
        if raised5 and self.normalize_note_name(raised5) in norm_song:
            # strong indicator of the relative harmonic minor
            return DATABASE["MINOR_KEYS"][key]

        # otherwise keep major
        return key


    def get_chord_candidates(self, key):

        chord_candidates = {}

        if not key:
            return {}

        if key.endswith('m'):
            root = key[:-1]
            mode = 'minor'
        else:
            root = key
            mode = 'major'

        if mode == 'major' and root in DATABASE["CHORDS_BY_KEY"]:
            for chord in DATABASE["CHORDS_BY_KEY"][root]:
                chord_candidates[chord] = DATABASE["CHORD_NOTES"].get(chord, [])
            return chord_candidates

        if mode == 'minor':
            minor_key_name = f"{root}m"

            if minor_key_name in DATABASE["CHORDS_BY_KEY"]:
                for chord in DATABASE["CHORDS_BY_KEY"][minor_key_name]:
                    chord_candidates[chord] = DATABASE["CHORD_NOTES"].get(chord, [])

        return chord_candidates

    def chord_degree(self, chord_name, key_name):
        if not key_name:
            return None

        if key_name in DATABASE["CHORDS_BY_KEY"]:
            scale = DATABASE["CHORDS_BY_KEY"][key_name]
        else:
            rel_major = None
            for maj, minv in DATABASE["MINOR_KEYS"].items():
                if minv == key_name:
                    rel_major = maj
                    break
            if rel_major and rel_major in DATABASE["CHORDS_BY_KEY"]:
                scale = DATABASE["CHORDS_BY_KEY"][rel_major]
            else:
                scale = DATABASE["CHORDS_BY_KEY"].get(key_name, None)
        if not scale:
            return None
        try:
            return scale.index(chord_name) + 1
        except ValueError:
            return None

    def get_bass_of_bar(self, bar_objs):
        if not bar_objs:
            return None
        lowest = None
        for el in bar_objs:
            try:
                if isinstance(el, music21.chord.Chord):
                    for p in el.pitches:
                        if lowest is None or p.midi < lowest.midi:
                            lowest = p
                elif isinstance(el, music21.note.Note):
                    p = el.pitch
                    if lowest is None or p.midi < lowest.midi:
                        lowest = p
            except Exception:
                continue
        return self.normalize_note_name(lowest.name) if lowest else None

    def find_best_chord_for_bar(self, bar_note_names, bar_objs, chord_candidates, prev_chord=None, bar_index=0, total_bars=1):
        """
        Improved chord selection:
        - Weighted tone matches (root/3rd/5th)
        - Duration-weighted contributions
        - Bass/inversion bonuses
        - Top-note (melody) bonus
        - Short-note (likely passing tone) penalties
        - Progression / cadence bonuses
        """
        print(f"\n================ BAR {bar_index+1} ================")
        print("Notes:", bar_note_names)
        print("Prev chord:", prev_chord)
        print("-----------------------------------------------")

        try:
            if bar_objs:
                for el in bar_objs:
                    # music21 chord symbol (common when chords are written as symbols)
                    if isinstance(el, music21.harmony.ChordSymbol):
                        # figure e.g. 'F', 'Am', etc.
                        figure = getattr(el, 'figure', None)
                        if figure:
                            norm_fig = figure.strip()
                            print(f"Manual chord symbol found in bar: {norm_fig} -> using it.")
                            return norm_fig

                    # custom user annotation (if you attach 'manual_chord' attribute to an element)
                    if hasattr(el, 'manual_chord'):
                        mc = getattr(el, 'manual_chord')
                        if mc:
                            print(f"Manual chord annotation found in bar: {mc} -> using it.")
                            return mc
        except Exception:
            pass

        # Build a list of note occurrences with rough durations and midi for weighting.
        note_occurrences = []   # list of dicts: {'name':str, 'dur':float, 'midi':int}
        try:
            for el in (bar_objs or []):
                if isinstance(el, music21.note.Rest):
                    continue
                if isinstance(el, music21.note.Note):
                    name = self.normalize_note_name(el.pitch.name)
                    dur = getattr(el.duration, "quarterLength", 1.0) or 1.0
                    midi = getattr(el.pitch, "midi", None)
                    note_occurrences.append({'name': name, 'dur': float(dur), 'midi': midi})
                elif isinstance(el, music21.chord.Chord):
                    dur = getattr(el.duration, "quarterLength", 1.0) or 1.0
                    # for chords, add each pitch as separate occurrence
                    for p in el.pitches:
                        name = self.normalize_note_name(p.name)
                        try:
                            midi = p.midi
                        except Exception:
                            midi = None
                        note_occurrences.append({'name': name, 'dur': float(dur), 'midi': midi})
        except Exception:
            # fallback: use names only with default durations
            note_occurrences = [{'name': n, 'dur': 1.0, 'midi': None} for n in bar_note_names]

        # If bar_objs missing, infer from bar_note_names
        if not note_occurrences and bar_note_names:
            note_occurrences = [{'name': n, 'dur': 1.0, 'midi': None} for n in bar_note_names]

        total_note_dur = sum(n['dur'] for n in note_occurrences) or 1.0

        # Find top and bass note (by midi if available)
        top_note = None
        lowest_midi = None
        for n in note_occurrences:
            if n['midi'] is not None:
                if top_note is None or (n['midi'] > top_note['midi']):
                    top_note = n
                if lowest_midi is None or n['midi'] < lowest_midi:
                    lowest_midi = n['midi']
        # If midi not available, fallback to lexical top (not great but safe)
        if top_note is None and note_occurrences:
            # top as last note in the treble extraction order
            top_note = note_occurrences[-1]

        bass = self.get_bass_of_bar(bar_objs)
        print(f"(computed) bass from bar objects = {bass}")

        # 1) Base matching (original idea) but weighted by durations and improved weights
        scores = {}
        max_base = -1e9
        print(">> BASE SCORE (duration-weighted tone matches)")

        for chord, chord_notes in chord_candidates.items():
            # chord_notes is expected as list like [root, third, fifth] (may be missing)
            root = chord_notes[0] if len(chord_notes) > 0 else None
            third = chord_notes[1] if len(chord_notes) > 1 else None
            fifth = chord_notes[2] if len(chord_notes) > 2 else None
            extensions = chord_notes[3:] if len(chord_notes) > 3 else []

            base = 0.0
            # weights for the triad tones (these will be scaled by note duration fraction)
            tone_weights = {root: 3.0 if root else -0.5,
                            third: 2.0 if third else -0.5,
                            fifth: 2.0 if fifth else -0.5}

            # small weight for extensions
            ext_weight = 0.8

            for occ in note_occurrences:
                name = occ['name']
                dur = occ['dur']
                dur_factor = min(dur / total_note_dur * 4.0, 2.0)  # normalizing duration influence
                # reduce influence of very short notes (likely passing tones)
                if dur < 0.25:
                    dur_factor *= 0.45

                if name in tone_weights and tone_weights[name] > 0:
                    base += tone_weights[name] * dur_factor
                elif name in extensions:
                    base += ext_weight * dur_factor
                else:
                    # small penalty for extraneous notes (to avoid garbage chords)
                    base -= 0.4 * dur_factor

            # penalize if chord has zero intersection with bar notes (unlikely chord)
            intersection = set([n['name'] for n in note_occurrences]) & set([root, third, fifth] + extensions)
            if not intersection:
                base -= 2.0

            scores[chord] = {
                'base': base, 'root': root, 'third': third, 'fifth': fifth,
                'extensions': extensions
            }
            print(f"  {chord:<8} root={root:<3} third={third:<3} fifth={fifth:<3} -> base={base:.3f}")
            if base > max_base:
                max_base = base

        # 2) Candidate set with tolerance
        tol = 1.0  # slightly wider tolerance because of duration weighting
        candidates = [c for c, s in scores.items() if s['base'] >= max_base - tol]
        if not candidates:
            candidates = list(scores.keys())

        print("\n>> CHORDS IN TOLERANCE (max_base =", max_base, ")")
        for c in candidates:
            print("  ", c)

        # 3) strong-tone & bass scoring (improved weights & duration influence)
        print("\n>> STRONG + BASS SCORE (bass =", bass, ")")
        for c in candidates:
            s = scores[c]
            st = 0.0

            # count weighted occurrences of root/third/fifth across note_occurrences
            for occ in note_occurrences:
                name = occ['name']
                dur = occ['dur']
                dur_factor = min(dur, 2.0)
                if name == s['root']:
                    st += 3.0 * dur_factor
                elif name == s['third']:
                    st += 2.0 * dur_factor
                elif name == s['fifth']:
                    st += 1.5 * dur_factor
                elif name in s.get('extensions', []):
                    st += 0.8 * dur_factor

            # bass match bonus (strong cue)
            if bass and s['root'] and bass == s['root']:
                st += 5.0
            elif bass and s['fifth'] and bass == s['fifth']:
                st += 3.0
            elif bass and s['third'] and bass == s['third']:
                st += 2.0

            # top-note (melody) bonus — top note often signals chord tone
            top_bonus = 0.0
            if top_note:
                top_name = top_note['name']
                if top_name == s['root']:
                    top_bonus += 1.5
                elif top_name == s['third']:
                    top_bonus += 1.0
                elif top_name == s['fifth']:
                    top_bonus += 0.8
                elif top_name in s.get('extensions', []):
                    top_bonus += 0.6
            st += top_bonus

            scores[c]['strong'] = st
            print(f"  {c:<8} strong={st:.3f} (top_bonus={top_bonus:.2f})")

        # 4) progression/context scoring (keep but scaled)
        print("\n>> PROGRESSION SCORE")
        common_pairs = set()
        for prog in DATABASE.get("COMMON_CHORD_PROGRESSIONS", []):
            for i in range(len(prog) - 1):
                common_pairs.add((prog[i], prog[i+1]))

        for c in candidates:
            prog = 0.0
            if prev_chord:
                prev_deg = self.chord_degree(prev_chord, self.detected_key)
                curr_deg = self.chord_degree(c, self.detected_key)
                if prev_deg and curr_deg:
                    # stronger reward for strong cadences
                    if prev_deg == 5 and curr_deg == 1:
                        prog += 5.0
                    if prev_deg == 4 and curr_deg == 1:
                        prog += 2.5
                    if prev_deg == 5 and curr_deg == 6:
                        prog += 0.8
                    if (prev_deg, curr_deg) in common_pairs:
                        prog += 2.0
                    if abs(prev_deg - curr_deg) == 1:
                        prog += 0.6
                    if abs(prev_deg - curr_deg) > 3:
                        prog -= 0.8

            scores[c]['prog'] = prog
            print(f"  {c:<8} prog={prog:.3f}")

        # 5) special/cadential score (first/last bar etc.)
        print("\n>> SPECIAL (FIRST/LAST BAR / CADENCE)")

        for c in candidates:
            special = 0.0
            deg = self.chord_degree(c, self.detected_key)

            # first bar: tonic favored strongly, also allow common pickups
            if bar_index == 0:
                if deg == 1:
                    special += 8.0
                elif deg in (3, 5):
                    special += 1.5

            # last bar: strong tonic bias (cadence)
            if bar_index == total_bars - 1:
                if deg == 1:
                    special += 8.0
                elif deg == 5:
                    special += 4.0

            # penalize highly unlikely degrees for cadential positions (small)
            if bar_index == total_bars - 1 and deg in (2, 3):
                special -= 1.0

            scores[c]['special'] = special
            print(f"  {c:<8} special={special:.3f}")

        # 6) Non-harmonic / passing tone penalty (small)
        # If many short notes present, reduce score for chords that only match those short notes.
        short_notes = sum(1 for occ in note_occurrences if occ['dur'] < 0.25)
        short_penalty = short_notes * 0.4
        if short_notes:
            print(f"\n>> NON-HARMONIC INFO: short_notes={short_notes} penalty={short_penalty:.2f}")
        else:
            print("\n>> NON-HARMONIC INFO: none detected")

        # 7) total
        print("\n>> TOTAL SCORE")
        best = None
        best_total = -1e9
        for c in candidates:
            total = scores[c]['base'] + scores[c]['strong'] + scores[c]['prog'] + scores[c]['special']
            # subtract penalty if chord relies mainly on short notes (heuristic)
            # compute match_count_normal_long = total weight from notes >= 0.25 dur that match chord
            long_match_weight = 0.0
            for occ in note_occurrences:
                if occ['dur'] >= 0.25:
                    name = occ['name']
                    if name in [scores[c]['root'], scores[c]['third'], scores[c]['fifth']] or name in scores[c].get('extensions', []):
                        long_match_weight += min(occ['dur'], 2.0)
            # if long_match_weight is very small while base was positive, apply penalty
            if long_match_weight < 0.5 and scores[c]['base'] > 0:
                total -= 1.5

            # apply global short-note penalty proportionally if chord matches mostly short notes
            if short_notes:
                total -= short_penalty * (1.0 if long_match_weight < 1.0 else 0.2)

            scores[c]['total'] = total
            print(f"  {c:<8} total={total:.3f} (base={scores[c]['base']:.3f} strong={scores[c]['strong']:.3f} prog={scores[c]['prog']:.3f} special={scores[c]['special']:.3f})")

            if total > best_total:
                best_total = total
                best = c

        # 8) tie-breaker (prefer bass match, then strong, then base)
        ties = [c for c in candidates if abs(scores[c]['total'] - best_total) < 1e-9]
        if len(ties) > 1:
            ties.sort(key=lambda x: (
                # prefer exact bass root match
                1 if (bass and scores[x]['root'] and bass == scores[x]['root']) else 0,
                scores[x]['strong'],
                scores[x]['base']
            ), reverse=True)
            best = ties[0]

        print("\n>>> SELECTED CHORD:", best)
        print("===================================================")
        return best

    def analyze_chord_progression(self):
        print("=== MUSIC ANALYSIS ===")

        # Step 1: detect key
        detected_key = self.analyze_key_signature()
        print(f"Detected key from analysis: {detected_key}")

        if self.key_signature and self.key_signature != "None" and self.key_signature in DATABASE["MAJOR_KEYS"]:
            final_key = self.key_signature
            print(f"Using key signature from file: {final_key}")
        else:
            final_key = detected_key
            print(f"Using analyzed key: {final_key}")

        # check major/minor
        final_key = self.check_major_minor(final_key, self.all_notes)
        self.detected_key = final_key  # ensure stored

        # chord candidates (dict chord -> notes)
        chord_candidates = self.get_chord_candidates(final_key)
        print(f"Chord candidates for {final_key}: {list(chord_candidates.keys())}")

        # analyze per bar
        chord_progression = []
        prev_chord = None
        total_bars = len(self.bars_notes)

        print("\n=== BAR-BY-BAR ANALYSIS ===")
        for i, (bar_names, bar_objs) in enumerate(zip(self.bars_notes, self.bars_notes_objs)):
            best = self.find_best_chord_for_bar(bar_names, bar_objs, chord_candidates, prev_chord=prev_chord, bar_index=i, total_bars=total_bars)
            chord_progression.append(best)
            prev_chord = best

            chord_notes = chord_candidates.get(best, [])
            match_count = sum(1 for note in bar_names if note in chord_notes)
            print(f"Bar {i+1}: Notes {bar_names}")
            print(f"         -> Chord: {best} {chord_notes} (match: {match_count}/{len(bar_names)} notes)")

        # Roman numerals (attempt)
        if final_key in DATABASE["CHORDS_BY_KEY"]:
            chords_in_key = DATABASE["CHORDS_BY_KEY"][final_key]
        else:
            relative_major = None
            for major, minor in DATABASE["MINOR_KEYS"].items():
                if minor == final_key:
                    relative_major = major
                    break
            chords_in_key = DATABASE["CHORDS_BY_KEY"].get(relative_major, DATABASE["CHORDS_BY_KEY"]["C"])

        roman_numerals = []
        for ch in chord_progression:
            if ch in chords_in_key:
                pos = chords_in_key.index(ch) + 1
                roman = {1: "I", 2: "ii", 3: "iii", 4: "IV", 5: "V", 6: "vi", 7: "vii°"}.get(pos, "?")
                roman_numerals.append(roman)
            else:
                roman_numerals.append("?")

        print(f"\n=== FINAL RESULTS ===")
        print(f"Key: {final_key}")
        print(f"Chord Progression: {' - '.join(chord_progression)}")
        print(f"Roman Numerals: {' - '.join(roman_numerals)}")

        return chord_progression
```
